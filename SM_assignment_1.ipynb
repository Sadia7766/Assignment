```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sympy import Matrix
```

```python
A=np.random.rand(15)
B=np.random.rand(15)

print("Matrix 1 is:\n",A)
print("\nMatrix 2 is:\n",B)

A_flat = A.flatten()
B_flat = B.flatten()

x = np.arange(len(A_flat))
print(x)

plt.plot(x, A_flat, marker='o', linestyle='-', color='red', label="Matrix A")
plt.plot(x, B_flat, marker='s', linestyle='--', color='blue', label="Matrix B")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Graph of Matrix A and B")
plt.legend()
plt.grid(True)

plt.show()
```

```python
A=np.random.randint(1,10,(4,4))

print("Matrix 1:\n",A)

sns.heatmap(A, annot=True, cmap='coolwarm', linewidths=1.2)
plt.title('Heatmap of 4x4 Matrix')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
A=np.random.randint(1,10,(4,4))
B=np.random.randint(1,10,(4,4))

print("Matrix 1:\n",A)
print("\n Matrix 2:\n",B)

matrix_sum=A+B
matrix_diff=A-B
matrix_prod=np.dot(A,B)

print("\n Addition is:\n",matrix_sum)
print("\n Subtracton is:\n",matrix_diff)
print("\n Multiplication is:\n",matrix_prod)

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

axes[0].bar(range(1,17), matrix_sum.flatten(), color='skyblue')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('X values')
axes[0].set_ylabel('Y values')

axes[1].bar(range(1, 17), matrix_diff.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('X values')
axes[1].set_ylabel('Y values')

axes[2].bar(range(1, 17), matrix_prod.flatten(), color='brown')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('X values')
axes[2].set_ylabel('Y values')

plt.tight_layout()
plt.show()
```
